To get the trend of this frequency time series data- it is unclear if there is clear trend.
We can assess trends using Rolling statistics and Dickey-Fuller test.
We can plot the moving average and moving variance and see if it varies with time. At any point in time we can take the mean/average/variance. m is known as the window size.

```python
rolmean =month_mean.rolling(window = 8, center = False).mean()
rolstd = month_mean.rolling(window = 8, center = False).std()
```

```python
fig = plt.figure(figsize=(12,7))
orig = plt.plot(month_mean, color='blue',label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

We can clearly see that the mean is almost constant over time so we can reconfirm our conclusion that the time series is stationary and the original plot is not different from the Rolling mean plot. It is also clear to see from the plot that there is not much of a seasonal and trend effect hence it is stationary 

Next step is to perform the Dickey-Fuller test which is a statistical test for testing stationarity. The Null hypothesis for the test is that the time series is not stationarity.

```python
from statsmodels.tsa.stattools import adfuller

dftest = adfuller(month_mean)

# Extract and display test results in a user friendly manner
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dftest)
```

```python
print('Results of Dickey Fuller test:')
print(dfoutput)
```

The p-value is also well below the 0.05 threshold and we reject the null hypothesis as this is statistically significant and more than a statstical fluke

Since the test statistic is less than the critical value we can definitely say that we reject the null hypothesis and say that the series is stationary.

```python
import warnings
warnings.filterwarnings('ignore')
import itertools
import numpy as np
p = d = q = range(0,2)
pdq = list(itertools.product(p,d,q))
pdqs = [(x[0], x[1],x[2],12) for x in list (itertools.product(p,d,q))]
```

```python
ans =[]
for comb in pdq:
    for combs in pdqs:
        try:
            mod = sm.tsa.statespace.SARIMAX(month_mean, order =comb,
                                           seasonal_order = combs,
                                           enforce_stationarity = False,
                                           enforce_invertibility=False)
            output = mod.fit()
            ans.append([comb,combs,output.aic])
            print('ARIMA{} X {}12:AIC Calculated={}'.format(comb,combs,output.aic))
        except :
            continue
```

```python
# finding the parameters with MINIMAL AIC value
```

```python
ans_df = pd.DataFrame(ans, columns = ['pdq', 'pdqs', 'aic'])
ans_df.loc[ans_df['aic'].idxmin()]#this is the optimal option out of all the models we have considered
```

fitting the ARIMA time series model

```python
ARIMA_MODEL = sm.tsa.statespace.SARIMAX(month_mean, order =(0,0,0),
                                        seasonal_order=(0,1,1,12),
                                       enforce_stationarity=False,
                                       enforce_invertibility=False)
output = ARIMA_MODEL.fit()

print(output.summary())
```

The table above shows the coefficient of each feature and how it impacts the time series pattern. as the p-values of each coefficient is less than 0.05, they are all significant features of the model.

```python
output.plot_diagnostics(figsize =(15,7))
```

Getting the predictions for 01-01-2016 and calculate the confidence intervals

```python
pred  =output.get_prediction(start = pd.to_datetime('2016-01-01'), dynamic = False)
pred_conf = pred.conf_int()
```

```python
rcParams['figure.figsize'] = 15, 6
#Plot observed values
ax = month_mean['2015':].plot(label='observed')
#Plot predicted values
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.9)
#Plot the range for confidence intervals
ax.fill_between(pred_conf.index,
                pred_conf.iloc[:, 0],
                pred_conf.iloc[:, 1], color='g', alpha=.5)
#Set axes labels
ax.set_xlabel('Date')
ax.set_ylabel('Frequency')
plt.legend()
plt.show()
```

```python
freq_forecasted = pred.predicted_mean
freq_truth = month_mean['2016-01-01':]
# Compute the mean square error
mse = ((freq_forecasted - freq_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
# The Mean Squared Error of our forecasts is 0.15
```

The mean square error of our forecast is 0.1 which is close to MSE. An MSE of 0 is the ideal scenario but it is not typically possible
